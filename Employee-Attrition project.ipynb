## OPEN A FILE 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the dataset
df = pd.read_csv('Attrition data.csv')
df.head()
```

```python
df.tail()
```

```python
# Data cleaning and preprocessing
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Demographic analysis
plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition by Department')
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.countplot(x='JobRole', hue='Attrition',data=df)
plt.title('Attrition by Job Role')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='JobLevel', y='PerformanceRating', hue='Attrition', data=df)
plt.title('Attrition by Job Level and Performance Rating')
plt.show()
```

```python
# Performance analysis
plt.figure(figsize=(10, 6))
sns.countplot(x='YearsSinceLastPromotion', hue='Attrition', data=df)
plt.title('Attrition by Last Promotion')
plt.show()
```

```python
# Salary and benefits analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MonthlyIncome', y='StockOptionLevel', hue='Attrition', data=df)
plt.title('Attrition by Monthly Income and Stock Option Level')
plt.show()
```

```python
# Interactive dashboard using Plotly
fig = px.scatter(df, x='YearsAtCompany', y='MonthlyIncome', color='Attrition', hover_data=['Department', 'JobRole'])
fig.update_layout(title='Attrition Dashboard')
fig.show()
```
