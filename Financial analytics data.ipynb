```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Financial Analytics data.csv')

```

```python
df.head()
```

```python
df.tail()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()

# Calculate the correlation between Quarterly Sales and Market Capitalization
corr_coef = corr_matrix['Sales Qtr - Crore']['Mar Cap - Crore']
print('Correlation Coefficient:', corr_coef)

# Visualize the relationship between Quarterly Sales and Market Capitalization
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sales Qtr - Crore', y='Mar Cap - Crore', data=df)
plt.title('Relationship between Quarterly Sales and Market Capitalization')
plt.xlabel('Quarterly Sales (Crore)')
plt.ylabel('Market Capitalization (Crore)')
plt.show()
```
